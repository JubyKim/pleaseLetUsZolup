```python
address = []

for i,j in zip(df['의원명'], df['의원위치']) :
    address.append(f'{j} {i}')
address
```

```python
#googlemaps에서 위도와 경도 구하기
import googlemaps
gmaps_key = 'AIzaSyA27NvqeiA2mR6w3Bs1pi1FQOLD76tCASo' # API설정할 때 얻은 key 
gmaps = googlemaps.Client(key=gmaps_key)
lat = []
lng = []

for i in address :
    tmp= gmaps.geocode(i, language='ko')
    tmp_loc = tmp[0].get('geometry')
    lat.append(tmp_loc['location']['lat'])
    lng.append(tmp_loc['location']['lng'])
```

```python
df['위도'] = lat
df['경도'] = lng
```

```python
df.to_csv('Desktop/서울지역_예방접종_위도경도.csv', encoding ='cp949')
```
