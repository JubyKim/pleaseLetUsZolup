```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.common.action_chains import ActionChains
import time
import math

#Step 2. 사용자에게 검색 관련 정보들을 입력 받습니다.
print("=" *100)
print("예방접종 웹크롤러입니다.")
print("=" *100)

#Step 3. 크롬 드라이버 설정 및 웹 페이지 열기
driver = webdriver.Chrome() 

url = 'https://www.hira.or.kr/re/diag/getNewDiagNondeductibleYadmList.do?pgmid=HIRAA030009000000'
driver.get(url)
driver.maximize_window()
time.sleep(2)

driver.find_element_by_xpath('//*[@id="HC"]').click()
time.sleep(2)
driver.find_element_by_xpath('//*[@id="sidoCd"]/option[2]').click()
time.sleep(2)
driver.find_element_by_xpath('//*[@id="search-form"]/div[2]/p[3]/a').click()
time.sleep(10)
html = driver.page_source
soup = BeautifulSoup(html, 'html.parser')
vacCount = soup.find('strong','pointR').get_text()
vacCount = vacCount.replace(',','')
vacCount = int(vacCount)

hospName = []
typeOfVac = []
NameOfVac = []
priceOfVac = []
where = []
count = 0

requiredPageCount = math.ceil(vacCount/10)
print(f'{vacCount}건을 수집하기 위해 {requiredPageCount}페이지의 게시물을 조회합니다.')

for p in range(1, requiredPageCount+1): #현재 있는 페이지 위치
    time.sleep(2)
    html = driver.page_source
    soup = BeautifulSoup(html, 'html.parser')
    content= soup.find('table','tbl_default tbl_thbold mid ac').find('tbody').find_all('tr')
    time.sleep(2)
    for i in content :
        if i.find('td','tleft'): 
            hospName.append(i.find('td','tleft').get_text())
            typeOfVac.append(i.get_text().split('\n')[14])
            NameOfVac.append(i.get_text().split('\n')[15])
            priceOfVac.append(i.get_text().split('\n')[20])
            where.append(i.get_text().split('\n')[22])
        else : 
            hospName.append('')
            typeOfVac.append(i.get_text().split('\n')[12])
            NameOfVac.append(i.get_text().split('\n')[13])
            priceOfVac.append(i.get_text().split('\n')[18])
            where.append(i.get_text().split('\n')[20])
        count += 1 
        if count > vacCount:
                break
                   
    p+=1 #이동해야하는 페이지 위치.
    print(f'p : {p}')
    print(f'count : {count}')# 현재까지 우리가 모든 데이터의 갯수
    if p < 11 :
        driver.find_element_by_link_text(str(p)).click()
    elif  p == 11 :
        driver.find_element_by_xpath('//*[@id="container"]/div[1]/div[3]/div[2]/div[1]/div[3]/a[12]').click()
    else :
        if p % 10 != 1 :
            driver.find_element_by_link_text(str(p)).click()
        else :
            driver.find_element_by_xpath('//*[@id="container"]/div[1]/div[3]/div[2]/div[1]/div[3]/a[13]').click()
#     if p < 12 or p % 10 != 1 :
#         driver.find_element_by_link_text(str(p)).click()
# #     elif  p == 11: 
# #         driver.find_element_by_xpath('//*[@id="container"]/div[1]/div[3]/div[2]/div[1]/div[3]/a[11]').click()
#     if p >= 12 and p != 11 and p % 10 == 1 :
#         driver.find_element_by_xpath('//*[@id="container"]/div[1]/div[3]/div[2]/div[1]/div[3]/a[13]').click()
        
print("모든 작업이 종료되었습니다.")
```

```python
import pandas as pd
import numpy as np
```

```python
vaccineDF = pd.DataFrame()
vaccineDF['의원명'] = hospName
vaccineDF['병명'] = typeOfVac
vaccineDF['백신명'] = NameOfVac
vaccineDF['가격'] = priceOfVac
vaccineDF['의원위치'] = where
```

```python
vaccineDF_test = vaccineDF
```

```python
vaccineDF_test.isnull().sum()
```

```python
for i, hosp in enumerate(vaccineDF_test['의원명']):
    if hosp == '':
        vaccineDF_test['의원명'][i] = np.NaN
```

```python
vaccineDF_test = vaccineDF_test.fillna(method="ffill")
```

```python
vaccineDF_test.to_csv('서울지역_예방접종_utf8.csv', encoding = 'utf-8')
```
